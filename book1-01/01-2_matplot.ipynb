# 1장. 파이썬 기초

matplot 부분

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 6, 0.1)          # arange, linspace 검색해보자.
y = np.sin(x)

plt.figure(figsize=(4,4))
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0, 6)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(4,4))
plt.plot(x,y1, label='sin')
plt.plot(x,y2, label='cos', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
