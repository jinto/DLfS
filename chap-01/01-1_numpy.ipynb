```python
import numpy as np
```

```python
x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])
```

## 행렬의 사칙 연산

```python
print(x+y)   # 각 요소를 더한 행렬
print(x-y)   # 각 요소를 뺀 행렬
print(x*y)   # 곱한
print(x/y)   # 나눈
```

## 브로드캐스트
행렬의 각 요소에 연산을 한다

```python
print(y/2)   
```

```python
# 배열도 브로드캐스팅이 된다.
A = np.array([
    [1,2], 
    [3,4]
])
B = np.array(
    [1,2]
)
print(A*B)
```

<div></div>

<div></div>

```python
# 원소 접근
print(A)
print(A[0])
print(A[0,1])
```

```python
# 한줄로 만들기
C = A.flatten()
print(C)
```

```python
print(C[np.array([0,2])])   # 인자에 np.array를 주면, 여러 원소에 한꺼번에 접근 가능
print(C[np.array([True, True, False, False])])   # T,F의 배열을 줄 수 도있다.
```

```python
# 조건식을 줄 수 있다
print(C>2)
print(C[C>2])
```
