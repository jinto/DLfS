### 연쇄법칙과 계산 그래프

계산 그래프에서 맨 왼쪽의 $\frac{\partial{z}}{\partial{z}} \frac{\partial{z}}{\partial{t}} \frac{\partial{t}}{\partial{x}}$ 는 계산하고 나면 $\frac{\partial{z}}{\partial{x}}$ 가 된다. 즉 `x 가 변하면 z 가 얼마나 영향받는가` 하는 식이되고, 이것은 우리가 역전파로 얻고자하는 것과 동일하다.
