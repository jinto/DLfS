# 2장. 퍼셉트론

1957년(!) 프랑크 로젠블라크가 고안한 개념이다. (그랬구나!)

퍼셉트론 == 인공 뉴런 == 인공 신경 세포



1. 신호 2개($x_1$. $x_2$)를 받아서
2. 가중치들($w_1$, $w_2$)을 각각 곱해서
3. 특정한 임계값($theta$)를 넘는 경우에는 1, 아니면 0을 출력신호($y$)로 내보낸다.  ($theta$ 는 그림에 없음)

```python
import numpy as np
```

```python
# 각 논리회로는 진리표 대로 나오게 하면 되는 것. (AND는 TT 일때 T)
# theta 는 이 함수가 원하는 대로 동작하게 하는 값을 임의로 정한 것.
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
    
print(AND(0,0), AND(0,1), AND(1,0), AND(1,1))
```

```python
# 넘피 배열로 만들어도 동작한다.

def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(x*w) + b
    return int(tmp > 0)
    
print(AND(0,0), AND(0,1), AND(1,0), AND(1,1))


# OR 이나 NAND도 비슷하게 구현할 수 있다.
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum(x*w) + b
    return int(tmp > 0)

print(NAND(0,0), NAND(0,1), NAND(1,0), NAND(1,1))

def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(x*w) + b
    return int(tmp > 0)
    
print(OR(0,0), OR(0,1), OR(1,0), OR(1,1))   
```

## OR 게이트
위에서 만든 `OR` 게이트를 수식과 그림으로 보면 이렇다. (잘 안보이는데, 회색영역은 비스듬한 직선의 좌하단을 가리킨다)





## XOR 게이트

문제는 `XOR` 게이트.




논리회로를 여러개 쌓으면 되긴 된다. (XOR함수안에서 NAND, OR, AND를 부르면 된다.)




```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y

print(XOR(0,0), XOR(0,1), XOR(1,0), XOR(1,1))    
```

즉, 단층 퍼셉트론으로는 못만든다.

한층이 더 필요하다.

참고 : NAND 의 조합만으로 컴퓨터를 만들 수 있다. ('NAND에서 테트리스까지' 라는 책이 있다)
