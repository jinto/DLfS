```python
#    
import numpy as np
import matplotlib.pylab as plt

x = np.arange(-5.0, 5.0, 0.1)
y = np.exp(x)
plt.plot(x, y)
plt.show()
```

```python
#  
def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a - c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
a = np.array([0.3, 2.9, 4.0])
y= softmax(a)
print(y)

#  1     ->    .
print(np.sum(y))
```

```python
# softmax   .  . (     .)
a = np.array([4,6])
print(softmax(a))
print(sum(softmax(a)))
```
